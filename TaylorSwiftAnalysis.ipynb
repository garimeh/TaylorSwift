### Data Cleaning

```python
df['lyrics_tm'] = df['song_lyrics'].copy()
```

```python
import re
from numpy.random import randint

def clean_text(text):
    # remove \n
    text = text.replace('\n', ' ')
    # remove punctuation
    text = re.sub(r'[,\.!?]', '', text)
    #removing text in square braquet
    text = re.sub(r'\[.*?\]', ' ', text)
    #removing numbers
    text = re.sub(r'\w*\d\w*',' ', text)
    #removing bracket
    text = re.sub(r'[()]', ' ', text)
    text = text.replace('\\', '')
    # convert all words in lower case
    text = text.lower()
    return text
```

```python
df['lyrics_tm'] = df['lyrics_tm'].apply(clean_text)
```

```python
docs = df['lyrics_tm'].to_list()
docs[0]
```

```python
# characters, words, lines
df['#characters'] = df['lyrics_tm'].str.len()
df['#words'] = df['lyrics_tm'].str.split().str.len()
df['#uniq_words'] = df['lyrics_tm'].apply(lambda x: len(set(x.split())))
df['lexical_density'] = df['#uniq_words'] / df['#words']
```

```python
# most common phrases
from nltk import ngrams
def get_ngrams_from_series(series, n=2):
    lines = ' '.join(series).lower().split('\n')
    lgrams = [ngrams(l.split(), n) for l in lines]
    grams = [[' '.join(g) for g in list(lg)] for lg in lgrams]
    return [item for sublist in grams for item in sublist]
```

```python
# Top bi-grams
pd.Series(get_ngrams_from_series(df['lyrics_tm'], 2)).value_counts()[:20][::-1].plot(kind='barh')
```

```python
# Top tri-grams
pd.Series(get_ngrams_from_series(df['lyrics_tm'], 3)).value_counts()[:20][::-1].plot(kind='barh')
```

```python
# sentiment
import nltk
from nltk import sentiment
nltk.download('vader_lexicon')
```

```python
sentimenter = sentiment.vader.SentimentIntensityAnalyzer()
```

```python
sentimenter.polarity_scores(df['lyrics_tm'][0])
```

```python
bins = [-np.inf, -0.75, -0.25, 0, 0.25, 0.75, np.inf]
labels = ['heartbroken sad', 'emotional sad', 'wistful sad', 'wistful happy', 'emotional happy', 'feverish happy']
df['sentiment_score'] = df['lyrics_tm'].apply(lambda x: sentimenter.polarity_scores(x)['pos'] - sentimenter.polarity_scores(x)['neg'])
df['sentiment'] = pd.cut(df['sentiment_score'], bins=bins, labels=labels)
df
```
